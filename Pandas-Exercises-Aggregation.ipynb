# Aggregations

### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib
import seaborn as sns
from matplotlib import pyplot
```

### Step 2. Import the dataset occupation.csv from the folder or [here](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
users=pd.read_table("occupation.csv",sep="|")
```

### Step 3. Assign it to a variable called users.

```python
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby(by=["occupation"])[["age"]].mean().round(1).add_prefix("meanof_").head()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
users.gender=np.where(users.gender=="M",0,1)
```

```python
users.head()
```

```python
us=pd.read_table("occupation.csv",sep="|")
usc=pd.crosstab(us.occupation,us.gender,margins=True)
usc=usc.rename(columns={'All':'total'})
usc.head()
```

```python
usc['male_ratio']=round((usc.M/usc.total*100),1)
usc.sort_values(by='male_ratio',ascending=False)
usc.head()
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
u1=users.groupby(by=["occupation"])[["age"]].min().round(1).add_prefix("min_").head()
u2=users.groupby(by=["occupation"])[["age"]].max().add_prefix("max_")
pd.concat([u1,u2],axis=1)
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
u3=users.groupby(["occupation","gender"])[["age"]].mean().add_prefix("mean_")
```

```python
u3.pivot_table(index="occupation",columns='gender',values="mean_age").round(2).head()
```
